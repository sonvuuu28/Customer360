# **Mục lục**

**1. Chuẩn bị môi trường**<br>
- Import thư viện cần thiết<br>
- Tải các bảng dữ liệu đã khám phá từ file 01 hoặc CSV<br>

**2. Xử lý giá trị thiếu (Missing Values)**<br>
- Xác định các cột/bảng bị thiếu dữ liệu<br>
- Quyết định cách xử lý: xóa, điền giá trị hoặc đánh dấu<br>

**3. Xử lý dữ liệu trùng lặp**<br>
- Kiểm tra và loại bỏ dòng trùng lặp theo khóa chính hoặc tất cả cột<br>

**4. Chuẩn hóa kiểu dữ liệu (Data Types)**<br>
- Chuyển cột thời gian, số, categorical về kiểu chuẩn (mấy cột có giá trị quy định trước ví dụ: Hà Nội: HN, Gender: Male)<br>

**5. Chuẩn hóa dữ liệu categorical & text**<br>
- Chuẩn hóa tên cột, category, text<br>
- Kiểm tra consistency giữa các bảng<br>

**6. Tạo các cột tính toán/biến mới**<br>
- Thời gian xử lý đơn hàng, tổng tiền, số lượng sản phẩm…<br>

**7. Kiểm tra dữ liệu sau xử lý**<br>
- Kiểm tra số lượng dòng/cột, kiểu dữ liệu, missing values còn lại<br>
- Kiểm tra logic dữ liệu (vd: giao hàng < đặt hàng)<br>

**8. Lưu dữ liệu sạch**<br>
- Xuất CSV/Parquet để dùng cho file 03_EDA hoặc modeling<br>
